```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('bank-additional-full.csv', sep = ';')
X = df.drop('y',axis = 1)
```

```python
X_ohe = pd.get_dummies(X, columns = ['job','marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome'], drop_first = True)
X_train = X_ohe.iloc[:,:].values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(df['y'])
Y = Y.astype(float)
Y_train = Y
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
```

```python
# Feature Importance with Extra Trees Classifier
from sklearn.ensemble import ExtraTreesClassifier
# feature extraction
model = ExtraTreesClassifier()
model.fit(X_train, Y_train)
model_importance = model.feature_importances_
colnames = list(X_ohe.columns.values)
colnames = np.asarray(colnames)
colnames = np.reshape(colnames, (53,1))
model_importance = np.reshape(model_importance, (53,1))
final = np.append(colnames,model_importance,axis =1)
final_sorted = final[final[:,1].argsort()] 
```

## Now we print the feature importance

```python
final_sorted
```

# Here we see that 'age', 'euribor3m', 'campaign', 'emp.var.rate' and 'successful previous outcomes(poutcome_success)' are the 5 factors which have significant contribution to say yes

```python
print ("Nth Call \t Efficiency")
for i in range(1,25):
    goo = sum(dfohe.loc[dfohe['campaign']==i]['y_yes']) / float(dfohe.loc[dfohe['campaign'] >= i].shape[0])
    print(str((i))+" \t\t "+str(goo))
```

# after 7 calls we see that call efficiency drops below 1%

```python
print ("Age \t Efficiency")
for i in range(15,95):
    goo = sum(dfohe.loc[dfohe['age']==i]['y_yes']) / float(dfohe.loc[dfohe['age']>=i].shape[0])
    print(str((i))+" \t\t "+str(goo))
```

# Here we see that contacting elederly people with age greater than 50 gave a positive response.

```python
# Calculate how many calls were made in total
total_calls = sum(dfohe['campaign'])
print(total_calls)
```

## total calls made = 105754


```python
# Calculate how many calls were made after the 7th call
over_calls = sum(dfohe[dfohe['campaign']>7]['campaign']) - 7*dfohe[dfohe['campaign']>7].shape[0]
print(over_calls)

```

## Over Calls( After 7th Call) made = 9634

```python
# Calculate reduction in marketing cost
reduction=100*over_calls/total_calls
print(reduction)

```

# we get 9% reduction in marketing (calling) cost

```python
total_sales=float(dfohe[dfohe['y_yes']==1].shape[0])
print(total_sales)
```

```python
less_costly_sales=float(dfohe[(dfohe['campaign'] <= 7) & (dfohe['y_yes']==1)].shape[0])
print(less_costly_sales)
```

```python
sales_percent=100*less_costly_sales/total_sales
print(sales_percent)
```

## here we get 98.4% sales persistance

## The bank will get a 9% reduction in marketing cost if a person is called atmost 7 times, and also maintaining 98.4% sales.
